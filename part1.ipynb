```python
#5. Läs in en valfri CSV-fil från Kaggle och visa informatio om datan.

student_df = pd.read_csv("student_monitoring_data.csv")
student_df
```

```python
#6. Filtrera datan baserat på ett villkor och skapa ett cirkeldiagram.
# Datan är för jämnt fördelad för att kunna hitta några mönster. Fördelningen på attendance status med avseende på olika mood scores ger knappt ngn skillnad =(
max_mood_df = student_df[(student_df["Mood Score"] == 10)]
max_mood_df[["Attendance Status"]].groupby("Attendance Status").size().plot(kind="pie", autopct='%1.1f%%', title="Attendance Status Among Students With Max Mood Score")
```

```python
min_mood_df = student_df[(student_df["Mood Score"] == 1)]
min_mood_df[["Attendance Status"]].groupby("Attendance Status").size().plot(kind="pie", autopct='%1.1f%%', title="Attendance Status Among Students With Min Mood Score")
```
